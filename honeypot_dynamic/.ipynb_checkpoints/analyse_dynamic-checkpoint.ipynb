```python
import pickle
import numpy as np
dir = "/home/quanghuyngo/Desktop/CFR/Code/honeypot_dynamic/dygraph_blocking/dyMIP_dyadsim025_all_10000_10000_1_5_20_0_.pkl"

with open(dir, 'rb') as f:
    data_1 = pickle.load(f)
```

```python
score_list = data_1["average_score_list"]
n = 2
tmp_list = []
for i in range(int(len(score_list)/n)):
    index = i*n
    temp = []
    for j in range(index, index + n):
#         print(j)
        temp.append(score_list[j])
#     print("################")
#     tmp_list.append(np.std(temp))
    tmp_list.append(np.mean(temp))
tmp_list
```

```python
blocking_list = data_1["blocking_list"]
n = 2
tmp_list = []
for i in range(int(len(blocking_list)/n)):
    index = i*n
    temp = []
    for j in range(index, index + n):
#         print(j)
#         temp.append(score_list[j])
        temp.append(blocking_list[j])
#     print("################")
#     tmp_list.append(np.std(temp))
#     tmp_list.append(np.mean(temp))
#     print(temp)
    tmp_list.append(len(set.intersection(*map(set,temp))))
tmp_list
```

```python
def vote_blocking(budget, block_strategy):
    # ranking the nodes in block strategy 
    vote = dict()
    for strat in block_strategy:
        for i in strat:
            if i not in vote:
                vote[i] = 0
            else:
                vote[i] = vote[i] + 1
    print(vote)
    return sorted(vote, key=vote.get, reverse=True)[:]
```
